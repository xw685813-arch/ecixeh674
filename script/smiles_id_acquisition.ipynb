```python
import json

with open("compounds.json", "r", encoding="utf-8") as f:
    data = json.load(f)

def get_smiles_by_id(met_id):
    for metabolite in data:
        if metabolite.get("id") == met_id:
            return metabolite.get("smiles")
    return None 

met_id = "cpd00001"
smiles = get_smiles_by_id(met_id)


```

```python
import json
import pandas as pd

with open("compounds.json", "r", encoding="utf-8") as f:
    json_data = json.load(f)

id_to_smiles = {m["id"]: m.get("smiles", None) for m in json_data}

csv_data = pd.read_csv("metabolite_ids.csv")

csv_data["smiles"] = csv_data["id"].map(id_to_smiles)

csv_data.to_csv("metabolite_with_smiles.csv", index=False, encoding="utf-8-sig")

```
